```python
#hagamos una funcio que reempace los datos null con cero
def reemplazar_nulos_con_cero_en_columna(df, columna):
    df[columna] = df[columna].fillna(0)
    return df
```

```python
df_uno['revenue'].isnull().sum()
```

```python
df_uno['budget'].isnull().sum()
```

```python
#sustituyamos los valores nullos en las columnas 'revenue', 'budget' y replacemolos por cero
df_uno = reemplazar_nulos_con_cero_en_columna(df_uno, 'revenue')
df_uno = reemplazar_nulos_con_cero_en_columna(df_uno, 'budget')
```

```python
df_uno['budget'] = df_uno['budget'].astype(float)
df_uno['revenue'] = df_uno['revenue'].astype(float)
```

```python
# Crear la columna "return" utilizando apply() y una función lambda
df_uno['return'] = df_uno.apply(lambda row: row['revenue'] / row['budget'] if (row['revenue'] and row['budget']) and row['budget']!=0 else 0, axis=1)
```
